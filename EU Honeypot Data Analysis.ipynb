# Analysis of distributed honeypot data

#### Step 1: Initialise and import the required modules/libraries.

```python
%matplotlib inline
import pandas as pd
import json
import pylab
import sys
pylab.rcParams[figure.figsize] = (16.0, 5.0)
```

#### Step 2: Open the JSON file containing all of our exported MHN data and append then prepare each honeypot data channel.

```python
f = open('EU_honeypot_data.json')
count = 0
glastopf = []
kippo = []
dionaea = []
snort = []
for line in f:
    j = json.loads(line)
    temp = []
    temp.append(j["_id"]["$oid"])
    temp.append(j["ident"])
    temp.append(j["normalized"])
    temp.append(j["timestamp"]["$date"])
    temp.append(j["channel"])
    try:
        payload = json.loads(j["payload"])
        if j["channel"] == "glastopf.events":
            temp.append(payload["pattern"])
            temp.append(payload["filename"])
            temp.append(payload["request_raw"])
            temp.append(payload["request_url"])
            temp.append(payload["source"][0])
            temp.append(payload["source"][1])
            glastopf.append(temp)
        elif j["channel"] == "kippo.sessions":
            temp.append(payload["peerIP"])
            temp.append(payload["commands"])
            temp.append(payload["loggedin"])
            temp.append(payload["version"])
            temp.append(payload["urls"])
            temp.append(payload["hostPort"])
            temp.append(payload["session"])
            temp.append(payload["credentials"])
            temp.append(payload["commands"])
            kippo.append(temp)
        elif j["channel"] == "dionaea.connections":
            temp.append(payload["local_host"])
            temp.append(payload["remote_hostname"])
            temp.append(payload["connection_transport"])
            temp.append(payload["remote_port"])
            temp.append(payload["remote_host"])
            temp.append(payload["connection_type"])
            temp.append(payload["connection_protocol"])
            temp.append(payload["local_port"])
            dionaea.append(temp)
        elif j["channel"] == "snort.alerts":
            temp.append(payload["source_ip"])
            temp.append(payload["priority"])
            temp.append(payload["signature"])
            temp.append(payload["proto"])
            temp.append(payload["destination_port"])
            snort.append(temp)
        else:
            print j
    except Exception as e:
        print str(e)
f.close()
```

#### Step 3: The next step is to create a new dataframe for each of the honeypots and their respective data columns.

```python
glastopf_df = pd.DataFrame(glastopf, columns=['id','ident','normalized','timestamp','channel','pattern','filename','request_raw','request_url','attackerIP','attackerPort'])
kippo_df = pd.DataFrame(kippo, columns=['id','ident','normalized','timestamp', 'channel', 'peerIP','commands','loggedin','version','urls','hostPort','session','credentials', 'commands'])
dionaea_df = pd.DataFrame(dionaea, columns=['id','ident','normalized','timestamp','channel','local_host','remote_hostname','connection_transport','remote_port', 'remote_host', 'connection_type', 'connection_protocol', 'local_port'])
snort_df = pd.DataFrame(snort, columns=['id', 'ident', 'normalized', 'timestamp', 'channel', 'source_ip', 'priority', 'signature', 'proto', 'destination_port'])
```

#### Step 4: Clean up the data by converting the timestamp column from the data to a datetime data type.

```python
glastopf_df['timestamp'] = glastopf_df['timestamp'].apply(lambda x: str(x).replace('T', 'T '))
kippo_df['timestamp'] = kippo_df['timestamp'].apply(lambda x: str(x).replace('T', 'T '))
dionaea_df['timestamp'] = dionaea_df['timestamp'].apply(lambda x: str(x).replace('T', 'T '))
snort_df['timestamp'] = snort_df['timestamp'].apply(lambda x: str(x).replace('T', 'T '))
```

```python
glastopf_df['timestamp'] = pd.to_datetime(glastopf_df['timestamp'])
kippo_df['timestamp'] = pd.to_datetime(kippo_df['timestamp'])
dionaea_df['timestamp'] = pd.to_datetime(dionaea_df['timestamp'])
snort_df['timestamp'] = pd.to_datetime(snort_df['timestamp'])
```

```python
import GeoIP

gi = GeoIP.new(GeoIP.GEOIP_MEMORY_CACHE)

glastopf_df['attackerCountry'] = glastopf_df['attackerIP'].apply(lambda x: gi.country_name_by_addr(x))
kippo_df['attackerCountry'] = kippo_df['peerIP'].apply(lambda x: gi.country_name_by_addr(x))
dionaea_df['attackerCountry'] = dionaea_df['remote_host'].apply(lambda x: gi.country_name_by_addr(x))
snort_df['attackerCountry'] = snort_df['source_ip'].apply(lambda x: gi.country_name_by_addr(x))
```

#### Step 5: Place all of our attacker IP address from the various honeypots into a single array.

```python
glastopf_df['allattackerIP'] = glastopf_df['attackerIP']
kippo_df['allattackerIP'] = kippo_df['peerIP']
dionaea_df['allattackerIP'] = dionaea_df['remote_host']
snort_df['allattackerIP'] = snort_df['source_ip']

glastopf_df['allattackedPorts'] = glastopf_df['attackerPort']
kippo_df['allattackedPorts'] = kippo_df['hostPort']
dionaea_df['allattackedPorts'] = dionaea_df['local_port']
snort_df['allattackedPorts'] = snort_df['destination_port']
```

#### Step 6: Create a new dataframe that has some information which is common the the dataframes created above.

```python
cols = ['channel','timestamp','allattackerIP', 'allattackedPorts', 'attackerCountry', 'ident']

attacker_df = pd.DataFrame()
attacker_df = attacker_df.append(glastopf_df[cols], ignore_index=True)
attacker_df = attacker_df.append(kippo_df[cols], ignore_index=True)
attacker_df = attacker_df.append(dionaea_df[cols], ignore_index=True)
attacker_df = attacker_df.append(snort_df[cols], ignore_index=True)
```

#### Step 7: Reindex our data by timeseries/timestamp

```python
attacker_df = attacker_df.set_index('timestamp')
```

#### IP addresses which were detected across multiple honeypots indicating an Internet wide scan/attack.

```python
honeymap = {}
for ident in set(attacker_df.ident.tolist()):
    honeymap[ident] = {}
    honeymap[ident]['channel'] = list(set(attacker_df[attacker_df['ident'] == ident]['channel'].tolist()))[0]
    honeymap[ident]['ip'] = list(set(attacker_df[attacker_df['ident'] == ident]['allattackerIP'].tolist()))[0]
```

```python
for ip in set(attacker_df['allattackerIP'].tolist()):
    ids = set(attacker_df[attacker_df['allattackerIP'] == ip]['ident'].tolist())
    if len(ids) > 1:
        temp = []
        for i in ids:
            temp.append(honeymap[i]['ip'] + ':' + honeymap[i]['channel'])
        temp = set(temp)
        if len(temp) > 3:
            print "%s seen across %d honeypots (%s) with %d connections" %(ip, len(temp), ", ".join(temp), attacker_df[attacker_df['allattackerIP'] == ip].shape[0])
```
